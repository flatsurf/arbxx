# The C++ interface of exact-real
We create a module generated by two random real numbers:

```c++
#include <memory>
#include <iostream>
#include <exact-real/exact-real.hpp>
```

```c++
auto a = exactreal::RealNumber::random(exactreal::Arf(0), exactreal::Arf(1));
auto b = exactreal::RealNumber::random(exactreal::Arf(0), exactreal::Arf(1));
auto M = exactreal::Module<exactreal::IntegerRing>::make({a, b}, exactreal::IntegerRing{});
    
std::cout << *M << std::endl;
```

Some arithmetic in this module:

```c++
auto x = M->gen(0);
auto y = M->gen(1);

std::cout << x + y << " != " << x * y << std::endl;
```

Elements can be compared:

```c++
x * y == y * x
```

```c++
x * y < x + y
```

```c++
x * y < 1
```
